```python
import pandas as pd

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
df = pd.read_csv(url)

# Display first 5 rows
df.head()
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='FlightNumber', y='LaunchSite', hue='Class', palette='deep')
plt.title('Flight Number vs Launch Site')
plt.xlabel('Flight Number')
plt.ylabel('Launch Site')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PayloadMass', y='LaunchSite', hue='Class', palette='muted')
plt.title('Payload Mass vs Launch Site')
plt.xlabel('Payload Mass (kg)')
plt.ylabel('Launch Site')
plt.grid(True)
plt.show()
```

```python
# Calculate success rate per orbit
success_rate = df.groupby('Orbit')['Class'].mean().reset_index()
success_rate.columns = ['Orbit', 'SuccessRate']

plt.figure(figsize=(10, 6))
sns.barplot(data=success_rate, x='Orbit', y='SuccessRate', palette='viridis')
plt.title('Success Rate by Orbit Type')
plt.xlabel('Orbit Type')
plt.ylabel('Success Rate')
plt.xticks(rotation=45)
plt.ylim(0, 1.1)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='FlightNumber', y='Orbit', hue='Class', palette='coolwarm')
plt.title('Flight Number vs Orbit Type')
plt.xlabel('Flight Number')
plt.ylabel('Orbit Type')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PayloadMass', y='Orbit', hue='Class', palette='Set2')
plt.title('Payload Mass vs Orbit Type')
plt.xlabel('Payload Mass (kg)')
plt.ylabel('Orbit Type')
plt.grid(True)
plt.show()
```

```python
# Define function to extract year
def extract_year(dates):
    return [date.split("-")[0] for date in dates]

# Create a new 'Year' column instead of replacing 'Date'
df['Year'] = extract_year(df['Date'])

# Show result
df.head()
```

```python
# Calculate mean success rate per year
yearly_success = df.groupby('Year')['Class'].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_success, x='Year', y='Class', marker='o')
plt.title('Average Launch Success Rate by Year')
plt.xlabel('Year')
plt.ylabel('Average Success Rate')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Select features
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights',
               'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block',
               'ReusedCount', 'Serial']]

features.head()
```

```python
# One-hot encode the categorical columns
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])

# Display the result
features_one_hot.head()
```

```python
# Convert all values to float64
features_one_hot = features_one_hot.astype('float64')

# Confirm the data types
features_one_hot.dtypes.head()
```

```python
# Export to CSV file
features_one_hot.to_csv('dataset_part_3.csv', index=False)
```
